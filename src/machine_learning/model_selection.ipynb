```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, ShuffleSplit, cross_validate
```

```python
tic_tac_df = pd.read_csv("../../src/data/processed_tic_tac_results.csv")
tic_tac_df
```

```python
X = tic_tac_df.drop(columns=["CLASS"])
y = tic_tac_df["CLASS"]
```

## Logistic Regression

```python
logreg = LogisticRegression(penalty="elasticnet", solver="saga", max_iter=10000)


param_grid = {
    "C": np.linspace(0.1, 5, 10),
    "l1_ratio": np.linspace(0, 1, 5),
}

grid_search_logreg = GridSearchCV(
    logreg, param_grid, cv=5, scoring="accuracy", n_jobs=-1, verbose=1
)


outer_valid = ShuffleSplit(n_splits=1, test_size=0.20, random_state=2)

results_logreg = cross_validate(
    estimator=grid_search_logreg,
    X=X,
    y=y,
    cv=outer_valid,
    return_estimator=True
)

print(f"accuracy: {results_logreg['test_score']}")
print(f"best params: {results_logreg['estimator'][0].best_params_}")
```
